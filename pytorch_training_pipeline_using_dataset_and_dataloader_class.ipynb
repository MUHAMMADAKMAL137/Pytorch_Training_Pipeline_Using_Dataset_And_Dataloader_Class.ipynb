```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=["id", "Unnamed: 32"], inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
x_train, x_test, y_train, y_test=train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2)
```

```python
x_train.shape
```

```python
y_train
```

```python
scaler= StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
x_train.shape
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.fit_transform(y_test)
```

```python
y_train
```

```python
x_train_tensor=torch.from_numpy(x_train.astype(np.float32))
x_test_tensor=torch.from_numpy(x_test.astype(np.float32))
y_train_tensor=torch.from_numpy(y_train.astype(np.float32))
y_test_tensor=torch.from_numpy(y_test.astype(np.float32))
```

```python
x_train_tensor.shape
```

```python
from torch.utils.data import DataLoader, Dataset
class CustomeDataset(Dataset):
  def __init__(self, features, labels):
    self.features=features
    self.labels=labels
  def __len__(self):
    return len(self.features)
  def __getitem__(self, index):
    return self.features[index], self.labels[index]
```

```python
train_dataset=CustomeDataset(x_train_tensor, y_train_tensor)
test_dataset=CustomeDataset(x_test_tensor, y_test_tensor)
```

```python
train_dataset[3]
```

```python
train_loader=DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader=DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python
import torch.nn as nn
class Mysimplenn(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear=nn.Linear(num_features, 1)
    self.sigmoid=nn.Sigmoid()
  def forward(self, features):
    out=self.linear(features)
    out=self.sigmoid(out)
    return out
```

```python
learing_rate=0.1
epochs=25
```

```python
model=Mysimplenn(x_train_tensor.shape[1])
optimizer=torch.optim.SGD(model.parameters(), lr=learing_rate)
loss_function=nn.BCELoss()
```

```python
for epoch in range(epochs):
  for batch_featuers, batch_labels in train_loader:
    y_pred=model(batch_featuers)
    loss=loss_function(y_pred, batch_labels.view(-1,1))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  print(f"Epoch: {epoch+1}, Loss: {loss.item()}")


```

```python
model.eval()
accuracy_list=[]
with torch.no_grad():
  for batch_featuers, batch_labels in test_loader:
    y_pred=model(batch_featuers)
    y_pred=(y_pred>0.80).float()
    batch_accuracy=(y_pred==batch_labels.view(-1)).float().mean()
    accuracy_list.append(batch_accuracy)
overall_accuracy=sum(accuracy_list)/len(accuracy_list)
print(f"overal accuracy: {overall_accuracy:.4f}")
```
